```python
%%sql
-- get rowcounts for all tables in 
SELECT 
    schemaname,
    tablename as table_name,
    (xpath('/row/c/text()', query_to_xml('SELECT count(*) AS c FROM '||quote_ident(schemaname)||'.'||quote_ident(tablename), true, true, '')))[1]::text::int as row_count
FROM pg_catalog.pg_tables
WHERE schemaname = '{username}'  
ORDER BY row_count DESC;
```

```python
%%sql
-- 1. Take a look at first 25 records of campaign
SELECT * 
FROM campaign 
LIMIT 25;
```

```python
%%sql
-- 2. Transaction Granularity: Get transaction counts per store, showing data distribution
SELECT s.STORE_NAME, 
       COUNT(*) as transaction_count, 
       COUNT(DISTINCT t.CUSTOMER_NBR) as unique_customers,
       SUM(t.TOTAL_SALES) as total_revenue
FROM STORE s
JOIN TRANSACTION_POS t ON s.STORE_NBR = t.STORE_NBR
GROUP BY s.STORE_NAME;
```

```python
%%sql
-- 3. Email Campaign Effectiveness: Analyze click-through and purchase conversion rates
SELECT 
    CAMPAIGN_SHORT_DESC,
    COUNT(*) as emails_sent,
    COUNT(EMAIL_CLICK_DATE) as clicks,
    COUNT(PURCHASE_DATE) as purchases,
    COUNT(EMAIL_CLICK_DATE) * 100.0 / COUNT(*) as click_rate,
    COUNT(PURCHASE_DATE) * 100.0 / COUNT(*) as conversion_rate
FROM CAMPAIGN
GROUP BY CAMPAIGN_SHORT_DESC;
```

```python
%%sql
-- 4. Customer Demographics: Distribution of customers across income deciles
SELECT 
    z.INC_MEDIAN_HOUSEHOLD_INCOME_D10 as income_decile,
    COUNT(DISTINCT c.CUSTOMER_NBR) as customer_count
FROM CUSTOMER c
JOIN ZIP_DEMOGRAPHICS_DECILE z ON c.ZIP_CODE = z.ZIP
GROUP BY z.INC_MEDIAN_HOUSEHOLD_INCOME_D10
ORDER BY income_decile;
```

```python
%%sql
-- 5. Product Analysis: Most popular wine varietals by sales volume
SELECT p.VARIETAL, 
       SUM(t.QUANTITY) as total_units_sold,
       SUM(t.TOTAL_SALES) as total_revenue
FROM PRODUCT p
JOIN TRANSACTION_POS t ON p.SKU = t.SKU
GROUP BY p.VARIETAL
ORDER BY total_units_sold DESC;
```

```python
%%sql
-- 6. Data Quality: Check for missing values in critical fields
SELECT 
    COUNT(*) as total_records,
    COUNT(CUSTOMER_NBR) as valid_customer_records,
    COUNT(EMAIL_CLICK_DATE) as valid_click_records,
    COUNT(PURCHASE_DATE) as valid_purchase_records
FROM CAMPAIGN;
```

```python
%%sql
-- 7. Customer Purchase Patterns: Frequency and recency analysis
SELECT 
    c.CUSTOMER_NBR,
    COUNT(*) as transaction_count,
    MIN(t.TRANSACTION_DATE) as first_purchase,
    MAX(t.TRANSACTION_DATE) as last_purchase,
    SUM(t.TOTAL_SALES) as total_spent
FROM CUSTOMER c
JOIN TRANSACTION_POS t ON c.CUSTOMER_NBR = t.CUSTOMER_NBR
GROUP BY c.CUSTOMER_NBR
order by 4 desc
limit 50;
```

```python
%%sql
-- 8. Geographic Distribution: Customer concentration by region
SELECT 
    z.STATE,
    z.CITY,
    COUNT(DISTINCT c.CUSTOMER_NBR) as customer_count,
    COUNT(DISTINCT s.STORE_NBR) as store_count
FROM CUSTOMER c
JOIN ZIP z ON c.ZIP_CODE = z.ZIP
LEFT JOIN STORE s ON z.STATE = s.STORE_STATE
GROUP BY z.STATE,
        z.CITY
order by 3 desc;
```

```python
%%sql
-- 9. Campaign Product Performance: Which products generate the most engagement
SELECT 
    p.SKU,
    p.PRODUCER,
    p.VARIETAL,
    COUNT(DISTINCT cam.CUSTOMER_NBR) as customers_targeted,
    COUNT(cam.PURCHASE_DATE) as purchases_made,
    SUM(cam.PURCHASE_QTY * cam.UNIT_PRICE) as campaign_revenue
FROM PRODUCT p
JOIN CAMPAIGN cam ON p.SKU = cam.SKU
GROUP BY p.SKU, p.PRODUCER, p.VARIETAL;
```

```python
%%sql
-- 10. Price Point Analysis: Transaction distribution across price ranges
SELECT 
    CASE 
        WHEN t.UNIT_PRICE < 20 THEN 'Under $20'
        WHEN t.UNIT_PRICE < 50 THEN '$20-$50'
        WHEN t.UNIT_PRICE < 100 THEN '$50-$100'
        ELSE 'Over $100'
    END as price_range,
    COUNT(*) as transaction_count,
    SUM(t.TOTAL_SALES) as total_revenue
FROM TRANSACTION_POS t
GROUP BY CASE 
    WHEN t.UNIT_PRICE < 20 THEN 'Under $20'
    WHEN t.UNIT_PRICE < 50 THEN '$20-$50'
    WHEN t.UNIT_PRICE < 100 THEN '$50-$100'
    ELSE 'Over $100'
END;
```
